### 왜 시각화를 진행하는지
- 시각화 기능
- 데이터 분석의 나의 주장이나, 가설을 설득하기 위한 하나의 과정
- 나의 주장과 가설 등을 잘 뒷받침할 수 있는 시각화를 통해 상대방 설득하게 되는 것
- 시각화를 의미가 변수간의 관계 확인용도 있지만, 데이터 분석에서 본인의 주장이, 가설이 베이스로 진행이 되는 것

#### 시각화는 상대방을 쉽게 설득시킬 수 있어야 한다.
#### 데이터 시각화로 잘못되게 표현하면 안 된다.
#### 다양한 시각화 그래프가 많지만, 가장 단순하고 가장 확실한 눈에 이해할 수 있는 시각화 등이 중요, 화려하다고 좋은 게 아님
- 수치에 대한 주석처리와 숫자에 대한 정확한 표시는 기본이다.
- 이런 단위 부터해서 기본적인 수치에 대한 기준 명시하지 않으면 이 부분은 신뢰할 수 없다.

### 축의 범위 설정
- x축과 y축이 내가 원하는 범위의 값만 보여줄 수 있음
- 축의 범위를 조정 가능

- xlim, ylim

```python
plt.scatter([1,2,3,4],[5,6,7,8])
plt.xlim(0,3)
plt.ylim(5,7)
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
plt.scatter([1,2,3,4],[5,6,7,8])
plt.xlim(left = 10) 
plt.ylim(top=10)
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

## 눈금값 설정

### 범례설정
- legend 함수 사용

```python
plt.scatter([1,2,3,4],[5,6,7,8], label='x=4')
plt.xticks([1,2,3,4],['A','B','C','D'])
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

### plt문법이 겹치게 사용하면 추가 확장

```python
plt.scatter([1,2,3,4],[5,6,7,8], label='x=4')
plt.scatter([1,2,3,4],[8,7,6,5], label='x=3') #층이 쌓여질 수 있다.
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

### 두 가지 이상의 그래프를 그려보기

- 선그래프를 그려보기 plot 
- plt.plot() 안에 들어가는 파라미터는 조금씩 다를 수 있다.

```python
plt.figure(figsize=(15,10))
plt.plot([1,2,3],[4,5,6], label= 'x1')
plt.plot([1,2,3],[6,5,4], label= 'x2')
plt.legend()
```

### 색상, 선 스타일, 두께 설정 가능
- color,  linestyle, linewidth 조정할 수 있는 파라미터

```python
plt.figure(figsize=(15,10))
plt.plot([1,2,3],[4,5,6],color='red' ,label= 'x1')
plt.plot([1,2,3],[6,5,4],color='blue',label= 'x2')
plt.legend()
```

```python
plt.figure(figsize=(15,10))
plt.plot([1,2,3],[4,5,6],color='red',linewidth=8,linestyle='dotted',label= 'x1')
plt.plot([1,2,3],[6,5,4],color='blue',linewidth=3,label= 'x2')
plt.legend()
```

### 마커설정
- 해당 값에 대한 지정 marker
- . 점, o 원 s 사각형 * 별, D 다이아몬드 등등 
- marker , markersize 지정 가능

```python
plt.figure(figsize=(15,10))
plt.plot([1,2,3],[4,5,6],marker='*',markersize=10,color='red' ,label= 'x1')
plt.plot([1,2,3],[6,5,4],color='blue',label= 'x2')
plt.legend()
```

### 히트맵
- 2차원 배열로 각 요소들 색상으로 시각화하는 방법

```python
import pandas as pd
df = pd.DataFrame({'+score':[5,4,3],
                  '-score':[2,1,5]},
                 index = ['Python','R','SQL'])
```

```python
sns.heatmap(df)
```

- 데이터 표시하고 싶은 경우
- annot 인자 True
- fmt 사용하면 'd' 정수, 실수 등 표현 가능
- cmap 색상 조합을 다양하게 만들 수 있다.

```python
sns.heatmap(df, annot=True, fmt='.1f') # 정수 소수 값들 다 조정 가능
```

```python
sns.heatmap(df, annot=True, fmt='.1f', linewidth = 2, linecolor = 'black')
```

```python
sns.heatmap(df, annot=True, fmt='.1f',cmap='Greys', linewidth = 2, linecolor = 'black')
```

### 그래프를 실제로 축을 지정하여 그리기

```python
x1 = np.arange(1,10)
x2 = np.arange(1,30)
y1 = x1*2
y2 = x2*3
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,7))

#축을 지정해서 그래프를 그리기
ax[0].plot(x1,y1, color='red')
ax[1].bar(x2,y2, color='green')

ax[0].set_xlabel('ax[0] axis')
ax[1].set_xlabel('ax[1] axis')

ax[0].legend()
ax[1].legend()
```
