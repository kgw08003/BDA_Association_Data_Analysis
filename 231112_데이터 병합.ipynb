### Concat 함수
- 덩어리와 덩어리가 합쳐지는 것 
- 데이터프레임 끼리 서로 합치는 경우 (따로 어떤 join 조건 등이 없다. ) 
- axis 축 설정
- ignore_index 병합 후 인덱스 재설정 

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A':[1,2,3], 'B':[4,5,6]})
df2 = pd.DataFrame({'A':[4,5], 'B':[6,7]})
df3 = pd.DataFrame({'A':[7,8,9], 'B':[10,11,12],'C':[13,14,15]})
```

```python
display(df1)
display(df2)
```

```python
display(pd.concat([df1,df2], axis=0)) #0행, 1열 기준 
display(pd.concat([df1,df2], axis=1)) #0행, 1열 기준 
```

```python
display(pd.concat([df2,df1], axis=0)) #0행, 1열 기준 
display(pd.concat([df2,df1], axis=1)) #0행, 1열 기준 

# 가장 먼저 쓴 데이프레임이 기본 축 
```

```python
display(pd.concat([df2,df1], axis=0,ignore_index= True)) #0행, 1열 기준 
# 인덱스를 새롭게 정렬
```

```python
display(df1)
display(df2)
display(df3)
```

```python
display(pd.concat([df1, df2, df3], axis=0)) #0행, 1열 기준 
```

```python
display(pd.concat([df1, df2, df3], axis=1)) #0행, 1열 기준 
#중복컬럼이 생김
```

## merge 함수
- join이라는 함수가 들어감
- join 공통의 컬럼이 필요하다 ( pk 개념 )
- 공통적인 부분을 기준으로 합친다. 


-----
- merge 파라미터가 있지만 주요적으로는
- left : 가장 먼저 쓴 데이터프레임, 왼쪽에 있는 데이터프레임을 left
- right : 두 번째 쓴 데이터프레임, right (2개 기준일 때 3개이상으로 하게 되면 순서에 따라서 정렬)
- on : join key 사용할 컬럼을 적는 것!
- left_on : 왼쪽 데이터프레임에서 join 키로 사용할 컬럼 목록 적기
- right_on : 오른쪽 데이터프레임에서 join 키로 사용할 컬럼 목록 적기
- left_index : 왼쪽 데이터프레임의 join기준 인덱스 사용할지 여부
- right_index : 오른쪽 데이터프레임 join기준 인덱스 사용할지 여부
- how : join 방법을 결정한다. 기본 디폴트는 'inner'

```python
df1 = pd.DataFrame({'이름':['홍길동','박길동','김길동','이길동'], '반':['파문응','파문기','데분기','데분중']})
df2 = pd.DataFrame({'이름':['홍길동','박길동','오길동','이길동'], '점수':[100,50,80,90]})
#df3 = pd.DataFrame({'A':[7,8,9], 'B':[10,11,12],'C':[13,14,15]})
```

```python
display(df1)
display(df2)
```

- 조인을 하는 경우는 대부분
- 하나의 테이블, 데이터프레임에서 얻을 수 없는 결과들을 한 번에 테이블로 만들어서 다양한 데이터를 분석할 수 있음
- 반과 점수를 같이 합쳐서 길동이들의 반별 점수를 알 수 있음

```python
pd.merge(df1, df2, on='이름')
```

```python
pd.merge(df2, df1, on='이름')
```

- inner 조인 디폴트라-> 공통의 교집합만 출력

### join 키가 두 개 이상인 경우

```python
df1 = pd.DataFrame({'이름':['홍길동','박길동','김길동','이길동'], '반':['파문응','파문기','데분기','데분중'],'학번':[20,21,22,23]})
df2 = pd.DataFrame({'이름':['홍길동','박길동','오길동','이길동'], '점수':[100,50,80,90],'학번':[20,19,22,23]})
```

```python
display(df1)
display(df2)
```

```python
#2개 이상을 Join key로 사용

pd.merge(df1, df2, on=['이름','학번'])
```

```python
pd.merge(df1, df2, on=['학번','이름'])
```

### left_on, right_on
- 컬럼명이 다름
- 컬럼 이름은 다르지만 값이 같다면 그 내용을 지정해서 join

```python
df1 = pd.DataFrame({'이름':['홍길동','박길동','김길동','이길동'], '반':['파문응','파문기','데분기','데분중'],'번호':[20,21,22,23]})
df2 = pd.DataFrame({'이름':['홍길동','박길동','오길동','이길동'], '점수':[100,50,80,90],'학번':[20,19,22,23]})
```

```python
display(df1)
display(df2)
```

```python
#중복이 나오는 경우
pd.merge(df1, df2, left_on='번호', right_on='학번')
```

```python
#중복이 나오지 않음
pd.merge(df1, df2, left_on='이름', right_on='이름')
```

- join key로 잡은 것 외에 이름이라는 두 개의 컬럼이 동일
- 중복이 발생해서 이름_x, 이름_y ( 중복 )

### 조인의 다양한 방법 How
- how 인자는 조인 방법을 다양하게 설정할 수 있다.
- 'inner' - 공통의 교집한 컬럼을 가지고 거기에 교집합된 내용들만 병합하는 것 0
- 'left' - 왼쪽 데이터프레임의 join 키를 기준으로 병합 
- 'right' - 오른쪽 데이터프레임의 Join 키를 기준으로 병합
- 'outer' - 왼쪽과 오른쪽의 데이프레임의 조인키에 대한 합집합을 기준으로 병합

```python
df1 = pd.DataFrame({'이름':['A','B','C','D'], '반':['파문응','파문기','데분기','데분중']})
df2 = pd.DataFrame({'이름':['A','B','E','F'], '점수':[100,50,80,90]})
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, on='이름')
```

```python
pd.merge(df1, df2, on='이름', how='inner')
```

```python
pd.merge(df1,df2, on='이름',how='left')#left를 하면 df1의 모든 값은 고정으로 생성되고 그 옆에 df2 의 데이터 중 on컬럼과 같은 교집합만 붙는다 na발생할 수 있다.
```

```python
pd.merge(df1,df2, on='이름',how='right')
```

```python
display(df1)
display(df2)
```

- outer
- 전체 모든 값을 다 합친 후에 교집합 내용들만 데이터 기입하는 것

```python
pd.merge(df1,df2, on='이름',how='outer') #left, outer 모든 값을 다 합친 경우에 outer 만들어짐
```

## 샘플 데이터로 데이터 병합

```python
site = pd.read_csv('survey_site.csv')
survey= pd.read_csv('survey_survey.csv')
vit = pd.read_csv('survey_visited.csv')
```

```python
#3개를 한 번에 Join 상황이 아니라, 하나하나씩 join하여서 세 개를 하나의 테이블로 만들어야 하는 경우!
display(site)
#display(survey)
display(vit)
```

```python
# site, vit 합치기
# 어떤 컬럼을 기준으로 하고 left right inner 
# 해당 데이터의 로직과 비즈니스 도메인 등 데이터 베이스의 설계 개념을 이해야 함
# vit 전체 메타데이터 (전체를 만드는 유니크한 메타데이터)
# 메타스키마의 개념으로 전체를 관리하는 테이블이 항상 존재 

mg1 = pd.merge(site,vit, left_on=['name'], right_on=['site'])
```

```python
pd.merge(mg1,survey, left_on=['ident'], right_on=['taken'])
```
