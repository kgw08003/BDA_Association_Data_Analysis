## query()

### 파생변수
- 외부변수, 파생변수 
- 통상적으로는 내부 데이터를 가지고 새로운 변수를 만드는 것
- 재료 조건이 내부 데이터 안에 있고 그 데이터를 가지고 -> 새로운 변수를 만듬


- 그 변수를 만드는 이유
- 1. 데이터 분석에 필요한 경우 만들 수 있음. 새로운 평균이라는 기준점을 가지고 분석을 해야 하는 경우 (매출 CAGR 3개년치 성장율)
- 2. 모델링을 통해 피처에 대한 엔지니어링을 진행하는 경우
- 기타 등등..

```python
df['3_mean'] = (df['math']+df['english']+df['science'])/3 #평균 ( 수학, 영어, 과학의 평균)
```

```python
#assign 함수를 이용, 파생변수를 만들 수 있음

df.assign(mean_s = (df['math']+df['english']+df['science'])/3)
```

```python
df
```

### groupby 
- 그룹으로 묶음
- 그룹으로 통계치를 계산하거나, 카운팅을 계산하거나 등등
- 결국은 그룹으로 무엇을 묶는다는 것이 중요한 포인트!

- df.groupby(기준).agg(변수, 요약통계치) 결과를 확인

```python
df
df.groupby('nclass') #groupby 는 agg 함수랑 같이 많이 사용한다.
```

```python
df
```

```python
df.groupby('nclass').agg(ncl_math = ('math','mean'))
#nClass 별 수학 평균
```

```python
df.groupby('nclass').agg(ncl_math = ('math','sum'))
#nClass 별 수학 sum
```

```python
df.groupby('nclass').agg(ncl_math = ('math','std'),
                        ncl_eng = ('english','sum'),
                        ncl_math_mean = ('math','mean'))
```

```python
df_gp=df.groupby('nclass').agg(ncl_math = ('math','std'),
                        ncl_eng = ('english','sum'),
                        ncl_math_mean = ('math','mean'))
```

```python
df
```

```python
df_gp=df.groupby('nclass').agg(ncl_math = ('math','std'),
                        ncl_eng = ('english','sum'),
                        ncl_math_mean = ('math','mean'))
```

```python
df_gp=df.groupby('nclass').agg(ncl_math = ('math','std'),
                        ncl_eng = ('english','max'),
                        ncl_math_mean = ('math','std'))
```
