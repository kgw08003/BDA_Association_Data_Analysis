```python
#df.groupby('year').lifeExp.agg(원하는 함수를 넣어서 사용할 수 있음)
```

```python
#원하는 함수를 만들어서 넣을 수 있음
def my_mean(values):
    n = len(values)
    sum_1 = 0
    for value in values:
        sum_1+= value
    return sum_1/n
#원하는 함수를 만들어서 agg 같이 넣을 수 있음
```

```python
df.groupby('year').lifeExp.agg(my_mean)
```

- numpy를 이용해서 추가할 수 있음
- agg 함수에 추가

```python
import numpy as np
df.groupby('year').lifeExp.agg([np.mean, np.std, np.count_nonzero])
```

### merge 
- merge 데이터 데이터를 합치는 경우에 발생
- 실무에서는 DA sql 충분히 가능하지만,
- DS 데이터셋을 정리하는 경우도 있다.
- 테이블을 병합시키고 정리해서 원하는 테스트셋을 만드는 경우 
- pandas를 많이 사용

- 데이터간의 병합을 통해 원하는 새로운 테이블 만드는 개념
- merge
- concat

```python
test1 = pd.DataFrame({'class':['파문기','데분기','데분중','데분고'],
                    '인원':[100,200,300,40]})
test2 = pd.DataFrame({'class':['파문기','데분기','데분중','데분고'],
                    '벌점평균':[5,6,7,3]})
```

```python
print(test1)

print(test2)
```

- test1과 test2
- class에 따라 인원, 벌점평균을 같이 볼 수 있는 테이블 만드는 경우

- 공통의 키로 가지고 가서 class를 잡고 병합

```python
# how 조인의 방법 = left join, right join inner join, outer join
# on 공통의 컬럼
pd.merge(test1, test2, how='left', on='class')
```

```python
##concat
## 덩어리 + 덩어리 
## 공통적인 키를 잡는 게 아니라 테이블 테이블, 덩어리 덩어리 두 개가 붙는 개념

sp_1=pd.concat([test1, test2],axis=0) ## 공통적인 컬럼을 지정하지 않는다.
```

```python
sp=pd.concat([test1, test2],axis=1) ## 공통적인 컬럼을 지정하지 않는다.
```
