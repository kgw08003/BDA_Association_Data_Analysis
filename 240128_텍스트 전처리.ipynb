## 정규화 작업
- 동사 변화 되는 것
- 흔히 어간(stem) 변하지 않는 부분 이 어간을 추출한다는 개념
- 간다, 갔다 작다, 작으니, 작고 , 작아서 이런 것들 작다의 개념에서 파생되는 것

- 포터 스테머 정규화하는 패키지
- porterStemmer

```python
from nltk.stem import PorterStemmer #정규화 하는 패키지

stem = PorterStemmer()

print(stem.stem('go'), stem.stem('going'), stem.stem('goes'))
```

```python
tokens = word_tokenize(sp1)

#불용어 한 것처럼 반복문으로 진행
res1 = [stem.stem(token) for token in tokens]
```

```python
res1
```

## 표제어 추출 Lemmatization

- 표제어 추출의 경우는 주어진 단어를 기본형을 변환하는 것

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemma = WordNetLemmatizer()
```

```python
print(lemma.lemmatize('go'))
print(lemma.lemmatize('going'))
print(lemma.lemmatize('goes'))


print(lemma.lemmatize('going', pos ='v')) #품사를 지정해서 동사로 뽑을 수 있다.
```
